```python
file_name = "input/30.txt"
file_name1="input/40.txt"
with open('/home/boy/Downloads/project_linux/jiaoben/input/new.txt', 'r', encoding='utf-8') as file,open(file_name, 'w', encoding='utf-8') as file1,open(file_name1, 'w', encoding='utf-8') as file2:
    # 逐行读取
    for line in file:
        # 使用split方法按空白分割字符串
        parts = line.split("\t")
        # 提取第三个空白段前的内容
        if len(parts)<=1 or line=="..." or len(parts[1])>9 or len(parts[0])<9 or parts[1]==" " or len(parts[0])>9:
            #print(parts)
            continue
        result = ' '.join(parts[:2])
        # 去除空格和制表符
        result = result.replace(' ', '').replace('：', '').replace('\t', '').replace('\n', '')
        if(parts[0][0]=='3'):
            file1.write(result+'\n')
        if(parts[0][0]=='4'):
            file2.write(result+'\n')
        
            
```
